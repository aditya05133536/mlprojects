```python
scaler = StandardScaler()
df['Amount'] = scaler.fit_transform(df['Amount'].values.reshape(-1, 1))
```

```python
df = df.drop(['Time'], axis=1)
```

# 0>> means legit transaction
# 1>> means fradulent 


```python
legit= df[df.Class==0]
fraud= df[df.Class==1]
```

```python
legit.shape
```

```python
print(legit)
```

```python
fraud.shape
```

```python
legit.Amount.describe()
```

```python
df.groupby("Class").mean()
```

as this dataset is highly imbalanced we have to do under sampling for making it a balanced data 


simply we have to make a sample dataset containing similar distribution of normal and fradulent transaction 

fradulent transaction >> 492 


```python
legit_sample= legit.sample(n=492)
```

now concatenating two dataframes 

```python
new_df=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_df.head()
```

```python
new_df["Class"].value_counts()
```

```python
new_df.groupby("Class").mean()
```

train_test_split

```python
X= new_df.drop(["Class"],axis=1)
```

```python
Y= new_df["Class"]
```

```python
X_train, X_test, Y_train, Y_test= train_test_split(X,Y,test_size=0.2,stratify=Y,random_state=3)
```

```python
X.shape,X_train.shape,X_test.shape
```

Model_training

Logistic_regression


```python
model= LogisticRegression()
```

```python
model.fit(X_train,Y_train)
```

accuracy_score

```python
# accuracy on the training data 
```

```python
X_train_prediction= model.predict(X_train)
training_data_accuracy= accuracy_score(X_train_prediction,Y_train)
print("accuracy_score of the train data: ", training_data_accuracy)
```

```python
# accuracy score of the test data 
```

```python
X_test_prediction=model.predict(X_test)
test_data_accuracy= accuracy_score(X_test_prediction,Y_test)
print("accuracy of the test data: ", test_data_accuracy)
```

```python
with open('logistic_regression_fraud_model.pkl', 'wb') as model_file:
    pickle.dump(model, model_file)
```

```python
with open('scaler.pkl', 'wb') as scaler_file:
    pickle.dump(scaler, scaler_file)
```
